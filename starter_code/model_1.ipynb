```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python

df.shape
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]
selected_features = df[['koi_period', 'koi_time0bk', 'koi_slogg', 'koi_srad', 'koi_kepmag']]
df.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical

#assign x and y values
X = df.drop("koi_disposition", axis=1)
y = df["koi_disposition"]

#split training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=84, stratify=y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data

from sklearn.preprocessing import StandardScaler

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
from sklearn.svm import SVC 
model_1 = SVC(kernel='linear')
model_1.fit(X_train_scaled, y_train)
predictions = model_1.predict(X_test)
```

```python
print(f"Training Data Score: {model_1.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model_1.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid_1 = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid_1 = GridSearchCV(model_1, param_grid_1, verbose=3)
```

```python
# Train the model with GridSearch
grid_1.fit(X_train_scaled, y_train)
```

```python
print(grid_1.best_params_)
print(grid_1.best_score_)
```
